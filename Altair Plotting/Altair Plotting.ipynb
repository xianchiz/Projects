```python
import pandas as pd

wdi = pd.read_csv(
    "https://raw.githubusercontent.com/nickeubank/practicaldatascience"
    "/master/Example_Data/world-small.csv"
)
wdi.head()
```

# Exercise 1

```python
import statsmodels.formula.api as smf
results = smf.ols('polityIV ~ gdppcap08',
                   data=wdi).fit()
print(results.summary())
```

```python
wdi[['polityIV', 'gdppcap08']].corr()
```

# Exercise 2

Based on the results of this analysis, both the p-values of intercept and gdp per capita is lower than 0.05, so there shows the significance of linear regression between gdp per capita and polityIV, but the r square and correlation between two variables are not high.

# Exercise 6--log transformation

```python
import numpy as np
wdi["log_gdppcap08"] = np.log(wdi["gdppcap08"])
base1 = (alt.Chart(wdi).mark_point().encode(
x = alt.X('log_gdppcap08', scale=alt.Scale(zero=False)),
y = 'polityIV'
))

```

```python
fit = base1.transform_regression('log_gdppcap08', 'polityIV').mark_line()
base1 + fit
```

# Exercise 7

```python
loess = base1.transform_loess(
    "log_gdppcap08", "polityIV").mark_line(color = "red")
base1 + fit + loess
```

# Exercise 8

```python
base1 + fit + loess + alt.Chart(wdi).encode(
    x = alt.X('log_gdppcap08', scale = alt.Scale(zero=False)),
    y = alt.Y('polityIV', scale = alt.Scale(zero=False)),
    text = 'country'
).mark_text(size = 7)
```

```python
wdi13 = wdi.copy()
wdi13 
```

```python
wdi10
wdi10["origin"] = "non-big oil producers"
wdi10
```

```python
wdi13['origin'] = wdi13['country'].apply(lambda x: 'big oil producers' if x in countries else 'non-big oil producers')  
wdi13
```

```python
input_dropdown = alt.binding_select(options=["big oil producers", "non-big oil producers"], name = "Country")
selection = alt.selection_single(fields=["origin"], bind=input_dropdown)
color = alt.condition(selection,
                    alt.Color('origin:N', legend=None),
                    alt.value('lightgray'))
alt.Chart(wdi13, title = "GDP per Capita and Democracy Scores").mark_point().encode(
    x=alt.X('log_gdppcap08', scale=alt.Scale(zero=False)),
    y=alt.Y('polityIV', scale=alt.Scale(zero=False)),
    color=color,
    tooltip=['country', 'gdppcap08', 'polityIV']
).add_selection(
    selection
).interactive()
```
