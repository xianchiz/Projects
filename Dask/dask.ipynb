# Exercise 4

```python
ddf_ex4 = dd.read_csv(
    "arcos_2011_2012.tsv",
    sep="\t",
    dtype={
        "REPORTER_STATE": "object",
        "BUYER_STATE" : "object",
        "BUYER_NAME" : "object",
        "MME_Conversion_Factor" : "float64",
        "CALC_BASE_WT_IN_GM" : "float64",
        "REPORTER_NAME" : "object",
    },
    usecols = ['REPORTER_STATE', 'MME_Conversion_Factor', 'CALC_BASE_WT_IN_GM','REPORTER_NAME', 'BUYER_STATE', 'BUYER_NAME'],
)
```

```python
ddf_ex4["opioids"] = ddf_ex4["MME_Conversion_Factor"] * ddf_ex4["CALC_BASE_WT_IN_GM"]
ddf3 = ddf_ex4[['REPORTER_NAME', 'BUYER_STATE', 'opioids']]
ddf3.head(5)
```

```python
ex4= ddf3.groupby(["REPORTER_NAME", "BUYER_STATE"]).sum().reset_index()
ex4_1 = ex4.sort_values(by="opioids", ascending = False).drop_duplicates(subset = "BUYER_STATE", keep = "first")
ex4_1.compute()

```

the table shows respective companies which have the highest shippment in each states. 

```python
ex4_1['REPORTER_NAME'].value_counts().compute()
```

we can see that Mckesson corporation was a monopoly of morphine shipments in most of the states in the US over 2011 to 2012.
